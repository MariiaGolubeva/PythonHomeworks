```python
from operator import itemgetter
from itertools import dropwhile
from string import punctuation

def tokenize(s):
    """Break str s into a list of strings according to some procedure 
    tailored for the task at hand."""
    wthpunct = s.strip().split()
    words = []
    for w in wthpunct:
        words.append(w.strip(punctuation))
    
    for w in words:
        if w == '' or w == '\n':
            words.remove(w)
    
    #your code
    return words #specify the output of the function: the function returns the given text as a list of wordforms
```

```python
def gutenberg_file_wc(filename):
    """Open the file with name filename, tokenize it with your 
    tokenize function, and return a dict mapping words to their 
    counts in the file.  Tokenize only the material that occurs
    between the two lines given above."""
    alice = open(filename).read()
    a = alice.find("*** START OF THIS PROJECT GUTENBERG EBOOK") 
    z = alice.find("*** END OF THIS PROJECT GUTENBERG EBOOK") 
    alice_new = alice[a:z]
    alice_words = []
    print(tokenize(alice_new))
    for word in tokenize(alice_new):
        alice_words.append(word.lower())
    alice_dict = {}
    for word in alice_words:
        if word not in alice_dict:
            alice_dict[word] = alice_words.count(word)
    #insert your code here
    return alice_dict #the output of the function
print(gutenberg_file_wc('alice.txt'))
```

## Words with clusters
## Task: 
### try to search for (word) initial consonant clusters using regular expression(s) (Don't forget to import re at the beginning of your program) -- 3 points;
### This function takes a file and returns a list of all words with initial consonant clusters. 
#### I hope it works correct, but there seems to be too little of these words in Alice - only about 500.

```python
import re

def words_wth_clusters(filename):
    
    alice = open(filename).read()
    a = alice.find("*** START OF THIS PROJECT GUTENBERG EBOOK") 
    z = alice.find("*** END OF THIS PROJECT GUTENBERG EBOOK") 
    all_words = alice[a:z]
    striped = tokenize(all_words)
    two_consonants = []
    for word in striped:
        if re.search('\A[WwRrTtPpLlKkJjHhGgFfDdSsZzXxCcVvBbNnMm]{2}[a-zA-Z]+', word):
            two_consonants.append(word.lower())
    clusters_words = two_consonants.copy()
    for word in two_consonants:
        if re.search('\A[TtSsPp][Hh]+',word) or re.search('\A[Ww][RrHh]+',word) or re.search('\A[KkGg][Nn]+',word):
            clusters_words.remove(word)
        if word == 'two' or word == 'http://www.pgdp.net' or word == 'chorus':
            clusters_words.remove(word)
            
    return clusters_words
print(words_wth_clusters('alice.txt'))
```

## This function takes a list from the previous function and prints a dictionary of clusters and their counts, total amount of clusters and a list of all clusters sorted alphabetically  

```python
def clusters(words):
    clusters_list = []
    for word in words:
        clusters_list.append(word[:2])
    clusters_count = {}
    for word in clusters_list:
        if word not in clusters_count:
            if re.search('s[ck]',word):
                clusters_count['sk'] = clusters_list.count('sk')+clusters_list.count('sc')
            else:
                clusters_count[word] = clusters_list.count(word)
            
        
    print(clusters_count, '\n', 'Total number of clusters: ', len(clusters_count))
    print('Alphabetically sorted clusters: ', sorted(clusters_count.keys()))
    
    
clusters(words_wth_clusters('alice.txt'))
```
